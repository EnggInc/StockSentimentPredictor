```python
import pandas as pd
import datetime as dt
```

```python
headlines = pd.read_csv("data/headlines.csv")
```

```python
pricemovements = pd.read_csv("data/price_movement.csv")
```

```python
pricechange3 = pd.read_csv('data/pricechange3.csv')
pricechange5 = pd.read_csv('data/pricechange5.csv')
```

```python
pricechange3
```

```python
headlines
```

```python
#pricemovements['Date'] = pd.to_datetime(pricemovements['Date']).dt.date
pricemovements['date'] = pricemovements['Date'].str[:10]
```

```python
type(pricemovements.date)
pricemovements
```

```python
# turn into 3 columns
mixed = pd.melt(pricemovements,id_vars = ['date'])
mixed.columns = ['date', 'stock', 'is_up']
mixed['is_up'] = mixed['is_up'].astype(str)
mixed
```

```python
type(mixed.date)
```

```python
is_up = pd.merge(headlines, mixed, on=['date', 'stock'])
is_up
```

```python
is_up.to_csv('data/is_up.csv')
```

```python
# turn into 3 columns
mixed = pd.melt(pricechange3,id_vars = ['Date'])
mixed.columns = ['date', 'stock', 'is_up']
mixed['is_up'] = mixed['is_up'].astype(str)
mixed
```

```python
is_up3 = pd.merge(headlines, mixed, on=['date', 'stock'])
is_up3
```

```python
mixed = pd.melt(pricechange5,id_vars = ['Date'])
mixed.columns = ['date', 'stock', 'is_up']
mixed['is_up'] = mixed['is_up'].astype(str)
mixed['date'] = mixed['date'].str[:10]
mixed
```
