```python
# We import pandas, and yfinance to to import closing prices of stocks
import pandas as pd
import yfinance as yf
```

```python
# We read our tickers.csv data and pass that to a list "tickers"
tickers = pd.read_csv("data/tickers.csv")
tickers = tickers.Symbol.to_list()
```

```python
# We extract stock price data on the above tickers from yfinace
# Our headlines.csv have headlines from "2009-04-29" to "2023-01-12"
# We are going to use that time frame(a day before for the start date, a day after for the end)
df = yf.download(tickers, start = "2009-04-28", end="2023-01-11")
```

```python
# We want closig price as we are interested in if the stock closed high or low from the previous day
# We create a new data frame "df_close" for closing prices
df_close = df["Close"]
```

```python
# Let's inspect first few entries
df_close.head()
```

```python
# We calculate the percentage change for all closing prices in "df_close" dataframe 
# We create a new dataframe "df_change" to store percentage change
df_change = df_close.pct_change()
```

```python
# Let's inspect first few entries
df_change.head()
```

```python
# We create a fucntion to examine if the percentage change is negative or positive
# We return 0 if the change was negative, meaning the stock price fell
# We return 1 if the change was positive, meaning the stock price rose
# There seem to have no changes of price in 20 plus instances for many stocks
# So we return 2 if the change was zero, meaning the stock price stayed same
# We return 'Na' for NaN
def increased_decreased(change):
    if change < 0:
        return 0
    elif change > 0:
        return 1
    elif change == 0:
        return 2
    else:
        return "Na"
```

```python
# We create a new dataframe "df_inc_dec" that stores either of 0,1, or 2 indicating price movement
# It stores Na for NaN, we will deal with those later
df_inc_dec = pd.DataFrame()

# We iterate through all the tickers in the columns in "df_change"
# We apply our "increased_decreased()" function to each column
# And store the tickers as columns in our new dataframe "df_inc_dec"
for ticker in df_change.columns:
     df_inc_dec[ticker] = df_change[ticker].apply(increased_decreased)
```

```python
# Let's inspect first few rows in our new dataframe "df_inc_dec"
# The first row has 'Na' because we we did not have the previous day's price data
# We will delete the first row later
# We see 'Na' in some columns at the beginning because those stocks were not listed at S&P at that time
df_inc_dec.head()
```

```python
# Let's inspect last few rows
# We don't see 'Na' because the stocks are listed in S&P and we have their data
df_inc_dec.tail()
```

```python
# We drop the last row 
df_inc_dec = df_inc_dec.iloc[:-1 , :]
df_inc_dec.tail()
```

```python
# We drop the first row
df_inc_dec = df_inc_dec.iloc[1: , :]
df_inc_dec.head()
```

```python
# Now we save "df_inc_dec" as a csv file in data folder
df_inc_dec.to_csv('data/price_movement.csv')
```
